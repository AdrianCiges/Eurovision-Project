### Cargamos todas las tablas de datos

```python
import pandas as pd

df_GDP = pd.read_excel('../GPD_2002_2022.xlsx')
df_letras = pd.read_excel('../Letras_2002_2022.xlsx')
df_links = pd.read_excel('../Links_YouTube.xlsx')
df_puntos = pd.read_excel('../Puntos_2002_2022.xlsx')
df_influencia = pd.read_excel('../Ranking_Influencia_UE.xlsx')
df_reputacion = pd.read_excel('../Reputation_UE.xlsx')
df_shazam = pd.read_excel('../Shazam_Generos.xlsx')
df_spoty = pd.read_excel('../Spoty_2002_2022.xlsx')
df_youtube = pd.read_excel('../Youtube_Views_Likes_2002_2022.xlsx.xlsx')
```

### Creamos BBDD en SQL Workbench

```python
from sqlalchemy import create_engine
```

```python
str_conn='mysql+pymysql://root:pasword@localhost:3306' #Conectar al servidor

cursor=create_engine(str_conn) #Cursor que apunta al servidor

cursor.execute('drop database if exists ESC;') #Lanzar query en el servidor

cursor.execute('create database ESC;') #Crear BDD ESC
```

```python
str_conn='mysql+pymysql://root:pasword@localhost:3306/esc' #Conexión a BDD esc

cursor_db=create_engine(str_conn) #Cursor que apunta a BDD ESC
```

### Enviamos las tablas a nuestra BBDD

```python
df_GDP.to_sql(name='gdp', con=cursor_db, if_exists='replace', index=False)
```

```python
df_letras.to_sql(name='letras', con=cursor_db, if_exists='replace', index=False)
```

```python
df_links.to_sql(name='links', con=cursor_db, if_exists='replace', index=False)
```

```python
df_puntos.to_sql(name='puntos', con=cursor_db, if_exists='replace', index=False)
```

```python
df_influencia.to_sql(name='influencia', con=cursor_db, if_exists='replace', index=False)
```

```python
df_reputacion.to_sql(name='reputacion', con=cursor_db, if_exists='replace', index=False)
```

```python
df_shazam.to_sql(name='shazam', con=cursor_db, if_exists='replace', index=False)
```

```python
df_spoty.to_sql(name='spoty', con=cursor_db, if_exists='replace', index=False)
```

```python
df_youtube.to_sql(name='youtube', con=cursor_db, if_exists='replace', index=False)
```
