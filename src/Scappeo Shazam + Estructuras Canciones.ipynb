```python
estruc_resum = []

for i,estruc in enumerate(df_letras['estructura']):

    resum = []
    listeo = estruc.replace("'", '').strip('][').split(', ')
    
    for palabra in listeo:

        if palabra == 'Intro':
            resum.append('I')

        elif palabra == 'Verso':
            resum.append('V')

        elif palabra == 'Puente':
            resum.append('P')

        elif palabra == 'Estribillo':
            resum.append('E')

        elif palabra == 'Coda':
            resum.append('C')

        elif palabra == 'Interludio':
            resum.append('B')
        
        else:
            resum.append('UNKNOWN')

    string = ''.join(resum)
    estruc_resum.append(string)
    resum = []
    
estruc_resum
```

```python
df_letras['estruc_resum'] = estruc_resum
```
