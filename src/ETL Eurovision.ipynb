```python
anio_act = 2001
df = pd.DataFrame()

for link in tqdm(links_ESC): # Para cada link en la lista de links...
    
    driver.get(link) # Entra en la página...
    time.sleep(random.randint(4,8)) # Genero un número aleatorio para esperar entre 4" y 8" para que no me "pillen" como robot
    
    driver.execute_script("document.body.style.zoom='25%'")
    dur = driver.find_elements(By.CLASS_NAME, 'Type__TypeElement-goli3j-0.eDbSCl.Btg2qHSuepFGBG6X0yEN')
    duracion = [e.text for e in dur]

    canc = driver.find_elements(By.CLASS_NAME, 'Type__TypeElement-goli3j-0.gwYBEX.t_yrXoUO3qGsJS4Y6iXX.standalone-ellipsis-one-line')
    cancion = [e.text for e in canc]
    
    cant = driver.find_elements(By.CLASS_NAME, 'Type__TypeElement-goli3j-0.eDbSCl.rq2VQ5mb9SDAFWbBIUIn.standalone-ellipsis-one-line')
    cantante = [e.text for e in cant]
    
    anio_act += 1 # Actualizo el año en cada bucle
    anio = [anio_act for i in range(len(cancion))] # Creo una lista del año con el len de la lista cantante
    
    df_ESC = pd.DataFrame()
    df_ESC['anio'] = anio
    df_ESC['cancion'] = cancion
    df_ESC['cantante'] = cantante
    df_ESC['duracion'] = duracion
    
    df = pd.concat([df, df_ESC], axis=0)
    
    
df 
```
